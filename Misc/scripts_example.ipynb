### Behav_analysis.py

```python
import sys
sys.path.append('/home/c10370862/FEED_new/scripts/') #change this to location of scripts!
import behav_analysis
import time

#set parameters
sub_labels = ['01']
n_sessions = 4
task = 'expressive'
kwargs = dict(dual = False, class_weight = 'balanced')

#load data and create DFs
AU_dfs,emo_dfs = behav_analysis.decoding_data_prep(sub_labels,n_sessions,task)

#calculate decoding performance (both OHE methods to compare speed)
start_time = time.time()
score_1, auc_1, df_1 = behav_analysis.AU_decoding(AU_dfs,emo_dfs)
print("--- runtime decoding analysis 1: %s seconds ---" % (time.time() - start_time))
print(score_1)

start_time_2 = time.time()
score_2, auc_2, df_2 = behav_analysis.AU_decoding_OHE(AU_dfs,emo_dfs)
print("--- runtime decoding analysis 2: %s seconds ---" % (time.time() - start_time_2))
print(score_2)
#plot auc scores
behav_analysis.plot_auc_score([df_1[0],df_2[0]])

```

### pupil_analysis.py

```python
import sys
from pathlib import Path
sys.path.append('/home/c10370862/FEED_new/scripts/') #change this to location of script!
import pupil_analysis
from glob import glob
from os import path

main_path = str(Path('.').absolute().parent)
out = main_path + '/data/behav/ET/preproc' #tweak this to the data folder on your account

kwargs = {'basis_set_method':'fir', #can be changed to 'fourier'
         'interval': [0.0, 3.5] #interval for irf plot ([0,2.5] as default)
         }  
analysis_params = {
                'sample_rate' : 1000.0,
                'pupil_lp' : 6.0,
                'pupil_hp' : 0.01, #maybe change to 0.1
                'normalization' : 'psc',
                'regress_blinks' : True,
                'regress_sacs' : True,
                'regress_xy' : False,
                'use_standard_blinksac_kernels' : False,
                }
edfs_sub02 = glob('/srv/FEED/behav_data/sub-02/*irfmapper*.edf')

hdf5_sub_02 = []
msg_sub_02 = []
dfs_sub_02 = []
rfs_sub_02 = []

for edf in edfs_sub02:
    hdf5,msg = pupil_analysis.pupil_preproc(edf,out,**analysis_params)
    hdf5_sub_02.append(hdf5)
    msg_sub_02.append(msg)
    
    df, e, rf = pupil_analysis.pupil_response_mapper(hdf5,msg, **kwargs)
    dfs_sub_02.append(df)
    rfs_sub_02.append(rf)
```
